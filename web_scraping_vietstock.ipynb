```python
import requests
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from datetime import datetime
import pandas as pd
```

```python
error_list = []
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'}
PATH = "C:\\Users\\ADMIN\\Desktop\\Code\\Python\\DataScience\\stock_sentiment\\chromedriver\\chromedriver.exe"
options = webdriver.ChromeOptions()
options.add_argument('--incognito')
#options.add_argument('--headless')
```

```python
driver = webdriver.Chrome(PATH, options=options)
driver.get('https://vietstock.vn/chung-khoan/co-phieu.htm')
```

```python
unwanted_element = driver.find_element_by_class_name('trending-fixed')
driver.execute_script("""
   var unwanted_element = arguments[0];
   unwanted_element.parentNode.removeChild(unwanted_element);
""", unwanted_element)
```

```python
href=[]
get_page=100
```

```python
for page in range(get_page):
    try:
        html_txt = BeautifulSoup(driver.page_source, 'lxml')
        articles_title = html_txt.find_all('a',class_="fontbold",href=True)
        for link in articles_title:
            href.append('https://vietstock.vn'+link['href'])
        href = list(set(href))
        next_button = WebDriverWait(driver, 500).until(
            EC.presence_of_element_located((By.ID,'page-next '))
        )
        next_button.click()
        time.sleep(1)
    except Exception as e:
        error_list.append(e)
    finally:
        time.sleep(4)
        continue
driver.close()
```

```python
articles_content = []
articles_time = []
for i in range(len(href)):
    try:
        text = ''
        html_txt = BeautifulSoup(requests.get(href[i], headers=headers).content, 'lxml')
        find_time = datetime.strptime(html_txt.find('span', class_='date').text, '%d/%m/%Y %H:%M')
        article_html = html_txt.find_all('p', class_='pBody')
        if find_time == None or article_html == []:
            continue
        for i in range(len(article_html)):
            text+=article_html[i].text
        if len(text) < 1000:
            continue
        articles_time.append(find_time)
        articles_content.append(text)
        print(articles_time[-1])
        print(href[i])
        print("="*50)
    except Exception as e:
        error_list.append(e)
    finally:
        time.sleep(1)
        continue
```

```python
print('href:',len(href),'articles:',len(articles_content),'time_list:',len(articles_time))
```

```python
#Add dates as columns
csv_file_by_date = {}
for i in range(len(articles_time)):
    csv_file_by_date.update({str(articles_time[i].day)+'/'+str(articles_time[i].month)+'/'+str(articles_time[i].year):[]})
```

```python
#add article content by date
for article in range(len(articles_content)):
    time = str(articles_time[article].day)+'/'+str(articles_time[article].month)+'/'+str(articles_time[article].year)
    csv_file_by_date[time].append(articles_content[article])
```

```python
df = pd.DataFrame.from_dict(csv_file_by_date, orient='index')
df = df.transpose()
df = df.fillna('')
```

```python
col_name = []
for i in range(len(df)):
    col_name.append('Article '+str(i+1))
```

```python
df = pd.DataFrame.transpose(df)
df.columns = col_name
df.index.name = 'Date'
df.reset_index(inplace = True)
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date',ascending=False,ignore_index=True)
df
```

```python
#df.to_csv('vietstock_articles.csv', encoding='utf8', index=False)
df.to_csv('vietstock_articles_23_2m.csv', mode='a', encoding='utf8', index=False, header=False)
```
