```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'}
```

```python
href=[]
get_page=500
```

```python
for page in range(get_page):
    src = requests.get('https://vneconomy.vn/chung-khoan.htm?trang='+str(page+1), headers=headers).content
    html_txt = BeautifulSoup(src, 'lxml')
    articles_title = []
    for h3 in html_txt.find_all('h3', class_='story__title'):
        articles_title.append(h3.find('a',href=True))
    for link in articles_title:
        href.append('https://vneconomy.vn'+link['href'])
    href = list(set(href))
```

```python
for page in range(get_page):
    src = requests.get('https://vneconomy.vn/the-gioi-kinh-te.htm?trang='+str(page+1), headers=headers).content
    html_txt = BeautifulSoup(src, 'lxml')
    articles_title = []
    for h3 in html_txt.find_all('h3', class_='story__title'):
        articles_title.append(h3.find('a',href=True))
    for link in articles_title:
        href.append('https://vneconomy.vn'+link['href'])
    href = list(set(href))
```

```python
articles_content = []
articles_time = []
for i in range(len(href)):
    text = ''
    html_txt = BeautifulSoup(requests.get(href[i], headers=headers).content, 'lxml')
    find_time = html_txt.find('div', class_='detail__meta')
    article_html = html_txt.find_all('div', class_='detail__content')
    if find_time == None or article_html == []:
        continue
    for char in article_html[0].text:
        text+=char.replace('&nbsp','')
    if len(text) < 1000:
        continue
    articles_time.append(datetime.strptime(find_time.text, '%H:%M %d/%m/%Y'))
    articles_content.append(text)
    print(articles_time[-1])
    print(href[i])
    print("="*50)
```

```python
print('href:',len(href),'articles:',len(articles_content),'time_list:',len(articles_time))
```

```python
#Add dates as columns
csv_file_by_date = {}
for i in range(len(articles_time)):
    csv_file_by_date.update({str(articles_time[i].day)+'/'+str(articles_time[i].month)+'/'+str(articles_time[i].year):[]})
```

```python
#add article content by date
for article in range(len(articles_content)):
    time = str(articles_time[article].day)+'/'+str(articles_time[article].month)+'/'+str(articles_time[article].year)
    csv_file_by_date[time].append(articles_content[article])
```

```python
df = pd.DataFrame.from_dict(csv_file_by_date, orient='index')
df = df.transpose()
df = df.fillna('')
```

```python
col_name = []
for i in range(len(df)):
    col_name.append('Article '+str(i+1))
```

```python
df = pd.DataFrame.transpose(df)
df.columns = col_name
df.index.name = 'Date'
df.reset_index(inplace = True)
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date',ascending=False,ignore_index=True)
```
